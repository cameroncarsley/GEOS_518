### 1. Read in the dataset and perform any additional pre-processing needed (compute water years)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.tsa.api as sm
from pandas.core import datetools

# Load the Yellowstone River's discharge dataset
#df = pd.read_pickle('..\H:\Hydrologic_Modeling\YellowstoneRiverQ_2007-2017.pkl')
df = pd.read_pickle('YellowstoneRiverQ_2007-2017.pkl')

# Create a Water Year column for our pandas data frame. This is a pretty 
# simple thing to do, but may not be necessary if you're not dealing with
# discharge data. Here's how it goes:
# 1. Create an empty array that is full of zeros and equal in length to 
#    the number of days in the record
WY = np.zeros(len(df['Y'].index)) 

# 2. For those records where the month is less than 10, their associated 
#    year is the correct water year
WY[df['M'].values < 10] = df['Y'].values[df['M'].values < 10] 

# 3. For those records where the month is greater than or equal to 10, 
#    the correct water year is one more than the current calendar year
WY[df['M'].values >= 10] = df['Y'].values[df['M'].values >= 10] + 1

# 4. Save the water year array as a column in the pandas data frame, as an
#    integer
df['WY'] = WY.astype(int)

# Print the first and last 10 records just to make sure we loaded the data okay
qrows = np.concatenate([np.arange(0,10,1),np.arange(-11,-1,1)])
df.iloc[qrows,:]
```

### 2. Split the dataset into a "training" dataset used to estimate the parameters of the AR model, and a "test" dataset against which I will test the model.

```python
# In the following example, I segment the whole dataframe
# into a training dataset (everything that's not Water Year 2017) and
# a test dataset (everything that is Water Year 2017).
df_train = df[df.WY != 2017]  # != means does not equal
df_test  = df[df.WY == 2017]  # == means equals
```

## 3. Estimate the parameters of an AR(1) model using the training dataset

### a) Compute the autocorrelation function

```python
# AC with 1 day lag:

# Use the pandas .values operator to load the values of discharge (the 'Q' column)
# from the pandas dataframe into a numpy array
Q = df_train['Q'].values 

# Use numpy array indexing to get discharge from October 2, 2000 to September 30, 2015. 
# The 'end' of a numpy array can be indexed using -1: as if the index in the negative direction
# wraps back around to the end of the record.  Array is N-1 in length.
Qt   = Q[1:-1]   # make array of (t+1 to t-1)

# Similarly get discharge from October 1, 2000 to September 29, 2015 by getting data from the 
# beginning of the record to the 'end' minus 1 using the -2 as the index. Array is N-1 in length.
Qtm1 = Q[0:-2]   # make array of (t to t-2 )

# Use Matplotlib to create a scatter plot of these two time series, create a title and label axes
plt.figure(figsize=(14,10))  # make figure large enough to view easily
plt.plot(Qtm1,Qt,'o')       # plot (t+1 to t-1) vs (t to t-2)
plt.title('How correlated is today\'s discharge with yesterday\'s?')
plt.xlabel('$Q_{t-1}$ m$^3$/s')
plt.ylabel('$Q_t$ m$^3$/s')
plt.show()

R1 = np.corrcoef(Qt, Qtm1)
print('R value of today vs. 1 days ago:', R1[0,1])

rho1 = R1[0,1]
```

### 3 b)
### Use the Yule-Walker equations to estimate the AR(1) parameters

```python
# phi(1) = rho(1) = R1

phi1_AR1 = R1[0,1]   # Phi1 for AR(1) equals AR1 autocorrelation coeff.
print('AR(1) phi1 = ', phi1_AR1)

#AR1_a = np.mean(Q)   # a1 equals the mean of the data
#print('a1 =', AR1_a)

#print('Zhat(1) =', phi1_AR1,'*Ztm1 +', AR1_a )
```

### 3 c)
### Use the statsmodel AR tools to estimate the parameters based on the dataset (AR(1)).

```python
Qt = pd.Series(df_train['Q'].values,df_train['SerDates'].values)  # make a matrix of Q and Serial Dates

Qpacf = sm.pacf(Qt)                                               # partial autocorrelation function

plt.figure(figsize=(14,10))
plt.stem(Qpacf)                                                   # plot using a stem plot
plt.ylabel('Partial Autocorrelation Function (PACF) [-]',fontsize=16)
plt.xlabel('Lag Distance [day]',fontsize=16)
ax = plt.gca()
ax.tick_params('both',labelsize=16)
plt.show()

from statsmodels.graphics.tsaplots import plot_pacf
f, axarr = plt.subplots(1, 1, figsize=(14,10))
_ = plot_pacf(Qt,method='ols',lags=40,ax=axarr.axes)
axarr.set_title('Statsmodel plot of PACF',fontsize=16)
axarr.set_ylabel('Partial Autocorrelation Function (PACF) [-]',fontsize=16)
axarr.set_xlabel('Lag Distance [day]',fontsize=16)
axarr = plt.gca()
axarr.tick_params('both',labelsize=16)

```

## 3c
### Using statsmodel to fit AR(1)

```python
Qt = pd.Series(df_train['Q'].values,df_train['SerDates'].values) # Set Qt to test values

Q_AR1_model = sm.AR(Qt).fit(1)
print(Q_AR1_model.params)
```

### 3 d)
### Compare the YW parameters to those from the statsmodel library

```python
print('The YW parameter for the AR(1) model is:', phi1_AR1, ', while the statsmodel library parameter is 0.992168')
print('The YW parameter and the statsmodel library parameter are identical up to four decimal places, and have a difference of', (((phi1_AR1-0.992168)/phi1_AR1)*100.0),'%')
```

## 4.  Estimate the parameters of an AR(2) model using the training dataset
### a) Compute the autocorrelation function

```python
# AC with 2 days lag

Qt   = Q[2:-1] # start 2 days from now
Qtmk = Q[0:-(3)] # today

# Use Matplotlib to create a scatter plot of these two time series, create a title and label axes
plt.figure(figsize=(14,10))
plt.plot(Qtmk,Qt,'o')
plt.title('How correlated is discharge today with discharge 2 days ago',fontsize=16)
plt.xlabel('$Q_{t-40}$ m$^3$/s',fontsize=16)
plt.ylabel('$Q_t$ m$^3$/s',fontsize=16)
plt.show()

R2 = np.corrcoef(Qt, Qtmk)
rho2 = R2[0,1]
#print(rho2)
print('R value (rho 2) of today vs. 2 days ago:', R2[0,1])
```

### 4 b)
### Use the Yule-Walker equations to estimate AR(2) parameters

```python
# phi 1
phi1_AR2 = (1-((rho1**2.0-rho2)/(rho1**2.0-1)))*rho1  # YW equation for phi1
print('phi1_AR2 =', phi1_AR2)

# phi 2
phi2_AR2 = (rho1**2-rho2)/(rho1**2-1)                 # YW equation for phi2
print('phi2_AR2 =', phi2_AR2)
```

### 4 c)
### Use the statsmodel AR tools to estimate the parameters based on the dataset

```python
Qt = pd.Series(df_train['Q'].values,df_train['SerDates'].values) # Set Qt to test values

Q_AR2_model = sm.AR(Qt).fit(2)  # Use statsmodel AR function
print(Q_AR2_model.params)
```

### 4 d)
### Compare the YW parameters to those from the statsmodel library

```python
print('My calculated AR(2) YW parameters are phi1 =', phi1_AR2, 'and phi2 =', phi2_AR2, 'while the statsmodel library output phi1 = 1.505482 and phi2 = -0.517322')
print("The YW values were accurate compared to the statsmodel parameters to about the thousanths' place")
```

## 5. Apply the AR(1) and AR(2) models to the 'test' dataset

```python
Qtrain = df_train['Q'].values   # assign training values to vector
Qtest  = df_test['Q'].values    # assign testing values to vector

DatesTest = df_test['SerDates'].values  # assign dates to vector for plotting

Qttm1 = np.concatenate([Qtrain[-2:-1],Qtest[0:-1]])  
Qttm2 = np.concatenate([Qtrain[-3:-1],Qtest[0:-2]])

AR1_mu   = Q_AR1_model.params[0]  # assign white noise value for AR(1) model
AR1_phi1 = Q_AR1_model.params[1]  # assign phi1 for AR(1)

QhatAR1 = AR1_mu + AR1_phi1*Qttm1 # evaluate AR(1) model

AR2_mu   = Q_AR2_model.params[0]  # assign white noise for AR(2) model
AR2_phi1 = Q_AR2_model.params[1]  # assign phi1 for AR(2) model
AR2_phi2 = Q_AR2_model.params[2]  # assign phi2 for AR(2) model

QhatAR2 = AR2_mu + AR2_phi1*Qttm1 + AR2_phi2*Qttm2  # evaluate AR(2) model
```

## 6. Plot the modeled time series for the AR(1) and AR(2) models against the observed time              series for comparison

```python

plt.figure(figsize=(14,10))

plt.plot(DatesTest,QhatAR1,'b-')  # plot AR(1) model
plt.plot(DatesTest,QhatAR2,'g-')  # plot AR(2) model
plt.plot(DatesTest,Qtest,'r-')    # plot actual data
plt.ylabel('Discharge [m${}^3$/s]',fontsize=16)
plt.xlabel('Date',fontsize=16)
plt.legend(('AR(1)','AR(2)','Obs.'),fontsize=16)
ax = plt.gca()
ax.tick_params('both',labelsize=16)
plt.show()

# Zoom in to see minor differences in models vs. observation
plt.figure(figsize=(14,10))
plt.plot(DatesTest,QhatAR1,'b-')
plt.plot(DatesTest,QhatAR2,'g-')
plt.plot(DatesTest,Qtest,'r-')
plt.xlim([DatesTest[200],DatesTest[250]])  # zoomed period is water year days 200-250
plt.ylabel('Discharge [m${}^3$/s]',fontsize=16)
plt.xlabel('Date',fontsize=16)
plt.legend(('AR(1)','AR(2)','Obs.'),fontsize=16)
ax = plt.gca()
ax.tick_params('both',labelsize=16)
plt.show()
```

## Compare the predictions with observations

```python

# Compute the R^2 values for each prediction 
R2AR1  = np.corrcoef(QhatAR1,Qtest)**2
R2AR2  = np.corrcoef(QhatAR2,Qtest)**2
muAR1  = np.mean(QhatAR1-Qtest)
stdAR1 = np.std(QhatAR1-Qtest)
muAR2  = np.mean(QhatAR2-Qtest)
stdAR2 = np.std(QhatAR2-Qtest)

# Plot the AR(1) and AR(2) model results 
plt.figure(figsize=(14,10))
plt.subplot(121)
plt.plot(QhatAR1,Qtest,'bo')
plt.plot([20, 700], [20, 700], 'k-')
plt.title('AR(1) Model',fontsize=16)
plt.ylabel('Observed Discharge [m${}^3$/s]',fontsize=16)
plt.xlabel('Observed Discharge [m${}^3$/s]',fontsize=16)
ax = plt.gca()
ax.tick_params('both',labelsize=16)
ax.annotate('R${}^2$ = %.3f'%R2AR1[0,1], xy=(20,700), fontsize=16)
ax.annotate('Avg. error = %.3f'%muAR1, xy=(20,670), fontsize=16)
ax.annotate('Std. error = %.3f'%stdAR1, xy=(20,640), fontsize=16)

plt.subplot(122)
plt.plot(QhatAR2,Qtest,'go')
plt.plot([20, 700], [20, 700], 'k-')
plt.title('AR(2) Model',fontsize=16)
plt.xlabel('Observed Discharge [m${}^3$/s]',fontsize=16)
ax = plt.gca()
ax.tick_params('both',labelsize=16)
ax.annotate('R${}^2$ = %.3f'%R2AR2[0,1], xy=(20,700), fontsize=16 )
ax.annotate('Avg. error = %.3f'%muAR2, xy=(20,670), fontsize=16)
ax.annotate('Std. error = %.3f'%stdAR2, xy=(20,640), fontsize=16)

plt.show()
```
