```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# define parameters of the Moving Average process of order 2 (MA(2))
mu = 4.5
theta1 = 0.5
theta2 = 0.3
std_eps = 1.3

# define number of time steps
N = 25

# Generate time series of white noise with mean = 0, standard deviation (std_eps)
np.random.seed(6)  # tells computer to pick the same starting point
eps_t = np.random.normal(0.0, std_eps, N)

# plot the time series of white noise
plt.figure(figsize=(14, 10))
plt.plot(eps_t, 'x')
plt.show()
```

```python
mu = 4.5
theta1 = 0.5
theta2 = 0.3
std_eps = 1.3

# create a storage container for our MA process
Z1_t = np.zeros((N,1))
Z2_t = np.zeros((N,1))
T = np.arange(0, N, 1)

# Create the values of the nose process at timesteps t-1 and t-2 to initiate the model
eps_tm1 = np.random.normal(0.0, std_eps, 1)
eps_tm2 = np.random.normal(0.0, std_eps, 1)


for t in T:
    if t==0:
        Z1_t[t] = mu + theta1*eps_tm1 + theta2*eps_tm2
    if t == 1:
        Z1_t[t] = mu + theta1*eps_t[t-1] + theta2*eps_tm1
    if t > 1:
        Z1_t[t] = mu + theta1*eps_t[t-1] + theta2*eps_t[t-2]
   

for t in T:
    if t==0:
        Z2_t[t] = mu + theta1*eps_tm1 - theta2*eps_tm2
    if t == 1:
        Z2_t[t] = mu + theta1*eps_t[t-1] - theta2*eps_tm1
    if t > 1:
        Z2_t[t] = mu + theta1*eps_t[t-1] - theta2*eps_t[t-2]
   

        
        
plt.figure(figsize=(14, 10))
plt.plot(T, Z1_t, 'g-', label='Z1_t')
plt.plot( T, Z2_t, 'b--')
plt.show()
```
